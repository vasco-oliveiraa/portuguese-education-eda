```python
df.head()
```

```python
df.drop(["Unnamed: 2", "N.º ", "N.º .1", "N.º .9", "N.º .10"], axis=1, inplace=True)
```

```python
col_names = ["year", 
             "region", 
             "10_12_public_science_humanities", 
             "10_12_public_technology",
             "10_12_public_artistic", 
             "10_12_public_professional_CP",
             "10_12_public_professional_AC",
             "10_12_public_education",
             "10_12_public_vocational",
             
             "10_12_total_science_humanities", 
             "10_12_total_technology",
             "10_12_total_artistic", 
             "10_12_total_professional_CP",
             "10_12_total_professional_AC",
             "10_12_total_education",
             "10_12_total_vocational"            
] 
df.columns = col_names
```

```python
df["10_12_private_science_humanities"] = df["10_12_total_science_humanities"] - df["10_12_public_science_humanities"]

df["10_12_private_technology"] = df["10_12_total_technology"] - df["10_12_public_technology"]

df["10_12_private_artistic"] = df["10_12_total_artistic"] - df["10_12_public_artistic"]

df["10_12_private_professional_CP"] = df["10_12_total_professional_CP"] - df["10_12_public_professional_CP"]

df["10_12_private_professional_AC"] = df["10_12_total_professional_AC"] - df["10_12_public_professional_AC"]

df["10_12_private_education"] = df["10_12_total_education"] - df["10_12_public_education"]

df["10_12_private_vocational"] = df["10_12_total_vocational"] - df["10_12_public_vocational"]
```

```python
df.drop(["10_12_total_science_humanities", 
             "10_12_total_technology",
             "10_12_total_artistic", 
             "10_12_total_professional_CP",
             "10_12_total_professional_AC",
             "10_12_total_education",
             "10_12_total_vocational" ], axis=1, inplace =True)
```

```python
df.head()
```

```python
df["year"].fillna(method="ffill", inplace=True)
```

```python
df.set_index(["year", "region"], inplace=True)
```

```python
df.to_csv("alunos_ensino_secundario_por_oferta.csv")
```
