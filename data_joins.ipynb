```python
import pandas as pd
import os
```

```python
main_df = pd.read_csv("cleaned_files/alunos_por_escolaridade_clean.csv")
df_2 = pd.read_csv("cleaned_files/alunos_ensino_superior_clean.csv")
df_3 = pd.read_csv("cleaned_files/alunos_ensino_secundario_por_oferta.csv")
df_4 = pd.read_csv("cleaned_files/proporcao_CT_ensino_superior.csv")
df_5 = pd.read_csv("cleaned_files/students_per_computer.csv")
df_6 = pd.read_csv("cleaned_files/students_per_psychologist.csv")
```

```python
#joined_1 = main_df.join(df_2, on=["year", "region"], how="inner")
main = pd.merge(main_df, df_2, left_on=["year", "region"], right_on=["year", "region"], how="left")
```

```python
main = pd.merge(main, df_3, left_on=["year", "region"], right_on=["year", "region"], how="left")
```

```python
main = pd.merge(main, df_4, left_on=["year", "region"], right_on=["year", "region"], how="left")
```

```python
main = pd.merge(main, df_5, left_on=["year", "region"], right_on=["year", "region"], how="left")
```

```python
main = pd.merge(main, df_6, left_on=["year", "region"], right_on=["year", "region"], how="left")
```
